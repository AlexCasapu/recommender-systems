```python
# CONTENT-BASED FILTERING (CBF)
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Katalog mit ausführlicheren Keywords (Kombination aus Kategorie, Eigenschaften, Features)
products = pd.DataFrame({
    'Produkt': ['ThinkPad T14', 'MacBook Pro', 'Dell XPS', 'Gaming PC Xtreme', 'Office PC Basic'],
    'Keywords': [
        'laptop business windows leicht office programmieren robust',
        'laptop premium macos kreativ design programmieren leicht',
        'laptop premium windows design multimedia leicht',
        'desktop gaming windows leistungsstark schwer grafikkarte rgb',
        'desktop office windows günstig basic textverarbeitung'
    ]
})
products
```

```python
# Initialisierung des TFIDF Vektorisierers (ohne Stopwords für dieses Beispiel, da die Keywords bereits bereinigt sind)
tfidf = TfidfVectorizer() # Fix: Initialisierung hinzugefügt!

# Text in Vektoren umwandeln
tfidf_matrix = tfidf.fit_transform(products['Keywords'])

# Kosinus-Ähnlichkeit berechnen
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
sim_df = pd.DataFrame(cosine_sim, index=products['Produkt'], columns=products['Produkt'])
sim_df
```

```python
# Beispiel: Wenn ein Nutzer das 'ThinkPad T14' mag, was empfehlen wir ihm?
thinkpad_recommendations = sim_df['ThinkPad T14'].sort_values(ascending=False)[1:4] # Top 3 ohne das Item selbst
thinkpad_recommendations
```
