```python
# 1. COLLABORATIVE FILTERING (CF)
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Wir nutzen einen erweiterten Filmdatensatz (User-Item Rating Matrix)
data = {
    'Matrix': [5, 4, 1, 1, 3],
    'Inception': [4, 5, 2, 1, 4],
    'Titanic': [1, 1, 5, 4, 2],
    'Notebook': [2, 1, 4, 5, 1],
    'Avatar': [5, 4, 2, 2, 4]
}
ratings = pd.DataFrame(data, index=['Alice', 'Tom', 'Lisa', 'Max', 'John'])
ratings
```

```python
# Wer ist Alice am ähnlichsten? User Based CF
user_similarity = ratings.T.corr(method='pearson')
user_similarity
```

```python
# Alternativer Algorithmus: Welche Filme sind sich ähnlich, basierend darauf, WIE sie bewertet wurden? Item-Based CF
# Wir füllen NaN (falls vorhanden) mit 0 für die Cosine Similarity
ratings_filled = ratings.fillna(0)

# Cosine Similarity zwischen den Items (Spalten)
item_sim_matrix = cosine_similarity(ratings_filled.T)
item_similarity = pd.DataFrame(item_sim_matrix, index=ratings.columns, columns=ratings.columns)
item_similarity
```
