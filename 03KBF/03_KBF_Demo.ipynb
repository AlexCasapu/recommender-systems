```python
import pandas as pd

# Katalog mit Hotels
hotels = pd.DataFrame({
    'Hotel': ['Warmbaderhof', 'City Inn', 'Alpenblick', 'Strandresort'],
    'Sterne': [5, 3, 4, 4],
    'Preis_Pro_Nacht': [250, 80, 150, 180],
    'Spa': [True, False, True, False],
    'Sportangebot': [True, False, True, True]
})

# 1. Präferenzerhebung (Was ist das Bedürfnis des Nutzers?)
# Statt harter Constraints fragen wir nach dem Ziel des Nutzers.
nutzer_beduerfnis = "Günstiger Städtetrip"

# 2. Die Wissensbasis (Regelwerk)
# Hier liegt das Wissen des Systems: Wie übersetzt man das Bedürfnis in harte Filter?
def empfehle_basierend_auf_wissen(beduerfnis, katalog):
    if beduerfnis == "Entspannung und Gewichtsverlust":
        # Dafür braucht man ein Spa und Sportangebote
        return katalog[(katalog['Spa'] == True) & (katalog['Sportangebot'] == True)]

    elif beduerfnis == "Günstiger Städtetrip":
        # Hier zählt nur der Preis
        return katalog[katalog['Preis_Pro_Nacht'] <= 100]

    else:
        return katalog

# 3. Empfehlung generieren
empfehlungen = empfehle_basierend_auf_wissen(nutzer_beduerfnis, hotels)
print(f"Ihre Empfehlungen für das Bedürfnis '{nutzer_beduerfnis}':\n")
print(empfehlungen)
```
