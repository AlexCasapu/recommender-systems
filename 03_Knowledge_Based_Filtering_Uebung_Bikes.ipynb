```python
# KNOWLEDGE-BASED FILTERING (KBF)
import pandas as pd

# Wir bauen den Canyon Bike-Finder nach
# Die Wissensbasis (Katalog)
bikes = pd.DataFrame({
    'Modell': ['Aeroad CF SLX 7 Di2', 'Ultimate CF SLX 8', 'Grizl CF SL', 'Endurace CF 7', 'Speedmax CF'],
    'Kategorie': ['Road', 'Road', 'Gravel', 'Road', 'Triathlon'],
    'Untergrund': ['Glatt', 'Glatt', 'Mischung', 'Glatt', 'Glatt'],
    'Fahrstil': ['Racing', 'Racing', 'Trails', 'Komfort', 'Zeitfahren'],
    'Prioritaet': ['Aero', 'Allrounder', 'Allrounder', 'Komfort', 'Aero'],
    'Preisklasse': ['Privateer', 'Profi', 'Privateer', 'Privateer', 'Profi'],
    'Preis_Euro': [4799, 8499, 2999, 2500, 8999]
})
bikes
```

```python
# Interaktive Abfrage über Python input()
# Hinweis: Führen Sie die Zelle aus und tippen Sie Ihre Antwort in die Textfelder, die erscheinen.
print("Wähle Untergrund: [Glatt, Mischung]")
wahl_untergrund = input("Deine Wahl: ")

print("\nWähle Fahrstil: [Racing, Zeitfahren, Komfort, Trails]")
wahl_fahrstil = input("Deine Wahl: ")

print("\nWähle Prioritaet: [Aero, Allrounder, Komfort]")
wahl_prioritaet = input("Deine Wahl: ")

print("\nDeine Eingaben:", wahl_untergrund, "|", wahl_fahrstil, "|", wahl_prioritaet)
```

```python
# Das System leitet nun das perfekte "Match" ab (Mapping)
def finde_match(untergrund, fahrstil, prioritaet, df):
    match = df[
        (df['Untergrund'] == untergrund) &
        (df['Fahrstil'] == fahrstil) &
        (df['Prioritaet'] == prioritaet)
    ]
    return match

# Testen wir die Eingaben aus der Zelle zuvor
ergebnis = finde_match(wahl_untergrund, wahl_fahrstil, wahl_prioritaet, bikes)

print("\n--- IT'S A MATCH! Das ist das perfekte Bike für dich: ---")
ergebnis
```
